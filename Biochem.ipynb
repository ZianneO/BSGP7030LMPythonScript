# Loading in CSV


```python
url = "https://raw.githubusercontent.com/TheTrappist/teaching/main/Biochem6761/AU24/Sample_data/AU24_03_titration.csv"
data = pd.read_csv(url)
```

# Displaying head of the data

```python
data.head()
```

# Plotting Brutus' Tittration Dataset

```python
plt.plot(data['Equivalents_OH'], data['pH_Brutus'], label='Brutus', color='purple')
plt.xlabel('Equivalents of OH-')
plt.ylabel('pH')
plt.legend()
plt.show()
```

# Plotting Wolverine's Titration Dataset

```python
plt.plot(data['Equivalents_OH'], data['pH_Wolverine'], label='Wolverine', color='pink')
plt.xlabel('Equivalents of OH-')
plt.ylabel('pH')
plt.legend()
plt.show()
```

# Fitting the data to Henderson-Hasselbalch
## pH = pKa + log ([A-]/[HA]) 
## simplified pH = pKa
